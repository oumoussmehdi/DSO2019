```python
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_predict
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
pd.set_option('display.max_columns', 500)

X_train = pd.read_csv("data/X_train.csv", index_col=0, error_bad_lines=False)
X_test = pd.read_csv("data/X_test.csv", index_col=0, error_bad_lines=False)
X_train.head(3)
```

```python
y_train = pd.read_csv("data/y_train.csv", index_col=0)
```

```python
df = pd.concat([X_train,y_train], axis=1)
df.head()
```

```python
df_num = df.select_dtypes(include=[np.number])
## df._get_numeric_data()

print(df_num.shape)
df_num.head()
```

```python
#list(df.select_dtypes(include=[np.number]).columns.values)
features = ['images_count','image_width','image_height','year','product_width','product_length','shoe_size','product_height','weight','price']
```

### Modelisation

```python
from xgboost import XGBClassifier
clf = XGBClassifier(n_estimators = 10, 
                       learning_rate = 0.05,
                       max_depth=6,
                       n_jobs=1,
                       alpha = 0.1)

import numpy as np
X_train = X_train._get_numeric_data()
y_train = y_train.values
clf.fit(X_train, y_train, verbose=False)
```

### Prediction

```python
pred_cv = cross_val_predict(clf, 
                            X_train.loc[:, features], 
                            np.ravel(y_train),
                            method='predict_proba', 
                            cv=5, 
                            n_jobs=-1)
```

### Evaluation

```python
from sklearn.metrics import log_loss 

pred_train = clf.predict_proba(X_train)

print("LogLoss on train sample:",log_loss(y_pred=pred_train, y_true=y_train)) 
print("LogLoss on train sample (CV):",log_loss(y_pred=pred_cv, y_true=y_train))
```
